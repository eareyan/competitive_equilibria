```python
def compute_conf_interval(sample, confidence_level = 0.95):
    degrees_freedom = len(sample) - 1
    sample_mean = np.mean(sample)
    sample_standard_error = scipy.stats.sem(sample)
    lb, ub = scipy.stats.t.interval(confidence_level, degrees_freedom, sample_mean, sample_standard_error)

    return sample_mean, lb, ub, (ub - lb)/2
```

```python
def print_ci(data):
    sample_mean, lb, ub, width_ci = compute_conf_interval(data)
    #print(sample_mean,  lb, ub, width_ci)
    print(f"${sample_mean :.2e} \pm {width_ci :.2e}$ ")
```

```python
for eps in [1.25, 2.5, 5.0, 10.0]:
    print_ci(lsvm_clears[f"{eps}_um"])
```

```python
for eps in [1.25, 2.5, 5.0, 10.0]:
    print_ci(lsvm_not_clears[f"{eps}_um"])
```
