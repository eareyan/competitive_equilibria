```python
def simple_stats(total, ptable):
    results = pd.read_parquet(f"/Users/enriqueareyan/Documents/workspace/competitive_equilibria/experiments_results/markets_{total}.parquet")
    assert len(results) == results.index.nunique()
    how_many = 0
    # print(total)
    for i in [1, 0]:
        for j in [1, 0]:
            if total == 2:
                data = results[(results['linear_clears'] == i) & (results['quadratic_clears'] == j)]
            else:
                data = results[(results['linear_clears'] == i) & (results['quadratic_clears'] == j)]
            #print(f"{'yes' if i == 1 else 'no'}\t {'yes' if j == 1 else 'no'} \t\t {len(data):,}")
            if i == j == 1:
                total_yes = len(data)
            if i == j == 0:
                total_no = len(data)
            ptable.add_row([total, f"{len(results) : ,}", f"{'yes' if i == 1 else 'no'}", f"{'yes' if j == 1 else 'no'}", f"{len(data) : ,}"])
            how_many += len(data)
    #print(f"total_yes = {total_yes}, total_no = {total_no}")
    if total_no > 0:
        ptable.add_row(['', '', '', 'fraction no', f"{(total_no / len(results)) * 100.0 : .2f}"])
    #ptable.add_row(['----', '----', '----', 'total_total', f"{total : ,}"])
    ptable.add_row(['----', '----', '----', '----', '----'])
```

```python
ptable_stats = PrettyTable()
ptable_stats.field_names = ['size', 'how many', 'linear', 'quadratic', 'clear']
ptable_stats.align['clear'] = 'r'
total_total = 0
for i in range(2, 9):
    simple_stats(i, ptable_stats)
print(ptable_stats)
```
